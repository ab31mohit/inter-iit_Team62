```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
data = pd.read_csv('odometry_data.csv', header=None)
# Assign column names based on the provided nametags
data.columns = ['timestamp', 'u1', 'u2', 'f1', 'f2', 'f3', 'p_s', 'q_s', 'r_s', 
                'n_x', 'n_y', 'n_z', 'roll_rad', 'pitch_rad', 'yaw_rad', 'x', 'y', 'z']
```

```python

# Plotting
plt.figure(figsize=(12, 8))

# Plot each variable against timestamp
for column in data.columns[1:]:  # Skip the first column (timestamp)
    plt.plot(data['timestamp'], data[column], label=column)

plt.title('Odometry Data Plot')
plt.xlabel('Timestamp')
plt.ylabel('Values')
plt.legend()
plt.grid()
plt.show()
```

```python
import plotly.express as px
def interactive_plotter_csv(csv_file):
    num = -5
    try:
        while True:
            num = num-1 if int(csv_file[num-1]) else num
    except:
        pass
    try:
        file_number = int(csv_file[num:-4])
    except:
        file_number = 0
    data = pd.read_csv(csv_file, header=None)
    data.columns = ['timestamp', 'u1', 'u2', 'f1', 'f2', 'f3', 'p_s', 'q_s', 'r_s', 
                    'n_x', 'n_y', 'n_z', 'roll_rad', 'pitch_rad', 'yaw_rad', 'x', 'y', 'z']
    for column in data.columns[1:]:  # Skip the first column (timestamp)
        fig = px.line(data, x='timestamp', y=column, 
                    labels={'value': 'Values', 'variable': column},
                    title=f'Odometry Data for {column} from file number: {file_number}')
        fig.show()
```

```python
interactive_plotter_csv("odometry_data_46.csv")
```

```python
interactive_plotter_csv("odometry_data_45.csv")
```
